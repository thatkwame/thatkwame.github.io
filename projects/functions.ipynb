# <font color = Blue>Functions in Python<font>
- A block of code that only runs when you call it
- We use the keyword 'def' to define a function

```python
def number_squared(number):
    print(number**2)
```

```python
number_squared(5)
```

```python
def first_function():
    print("We made it!")
```

```python
first_function()
```

```python
def number_exponent(number,power):
    print(number**power)
```

```python
number_exponent(5,3)
```

### Arbitrary Arguments
- Here is when we do not want to specify the number of arguments in the function

```python
def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(5,6,1,2,3)
```

```python
args_tuple = (5,6,1,2,3)

def number_args(*number):
    print(number[0]*number[1])
```

```python
number_args(*args_tuple)
```

## Keyword Arguments
This can give you more control over the arguments, by simply specifying what each item in the funtion mean. With this you can change the position of the items in the argument without it affecting the function outcome

```python
def number_exponent(number,power):
    print(number**power)
```

```python
number_exponent(power= 5,number = 3)
```

## Arbitrary Keyword Arguments

```python
def number_kwarg(**number):
    print("My number is:" + number["integer"] + " " + number["integer2"])
```

```python
number_kwarg(integer = '07554', integer2 ="265442" )
```

# <FONT COLOR = blue>Converting Data Types<FONT>

For the conversion of integers, strings and floats use new_var_typ(var)

```python
car = "7"
type(car)
```

```python
conv_car = int(car)

type(conv_car)
```

## Converting lists, sets & tuples

```python
goal_list = [1,2,3,4]

type(goal_list)
```

```python
goal_tuple = tuple(goal_list)

type(goal_tuple)
```

```python
new_list = [1,2,3,2,1,2,3,4,3,2,1,2,2,3,2]

type(new_list)
```

```python
conv_to_set = set(new_list)


print(conv_to_set)
```

```python
type(conv_to_set)
```

For dictionaries we need to specify what we want to convert i.e. items (key, value), keys, values

```python
dict_type = {'name': 'Mark Jay', 'date of birth': 'September 21', 'Height': '176cm'}

type(dict_type)
```

```python
print(list(dict_type))
```

```python
print(list(dict_type.items()))
```

```python
print(list(dict_type.keys()))
```

```python
print(list(dict_type.keys()))
```

```python
type(str(dict_type))
```
